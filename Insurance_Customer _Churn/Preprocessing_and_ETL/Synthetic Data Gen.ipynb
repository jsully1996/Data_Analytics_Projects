```python
import pandas as pd
import zipfile
import os
```

## Read a CSV containing synthetic Data generated using Excel

```python
with zipfile.ZipFile(os.path.dirname(os.getcwd())+'/Data/Master_Dataset_1.zip') as zip:
    with zip.open('Master_Dataset_1.csv') as myZip:
        df1 = pd.read_csv(myZip) 
df1
```

## Using FAKER to generate a synthetic dataframe

```python
from faker import Faker
fake = Faker('en_IN')
#fake_df = pd.DataFrame(columns=['Name','Address','Phone_no','Job'])
fake_data = {"Customer ID":[],"Name":[], "Address":[],"Phone_no":[],"Job":[],"Company":[],"Credit Card Provider":[],"Email":[],"SSN":[]}
for i in range(len(df1)):
    fake_data['Customer ID'].append(fake.pystr_format(string_format='??#####', letters='ABCDEFGHIJKLMNOPQRSTUVWXYZ'))
    fake_data['Name'].append(fake.name())
    fake_data['Address'].append(fake.address())
    fake_data['Phone_no'].append(fake.phone_number())
    fake_data['Job'].append(fake.job())
    fake_data['Company'].append(fake.company())
    fake_data['Credit Card Provider'].append(fake.credit_card_provider())
    fake_data['Email'].append(fake.ascii_email())
    fake_data['SSN'].append(fake.ssn())
fake_df= pd.DataFrame(fake_data)
fake_df
```

### Remove Duplicates in the "Customer ID" column for both dataframes

```python
print("Before dropping duplicate Customer ID : {} values for Dataframe1".format(len(df1)))
print("Before dropping duplicate Customer ID : {} values for Dataframe2".format(len(fake_df)))
df1.drop_duplicates(subset=['Customer ID'], keep='first',inplace=True)
fake_df.drop_duplicates(subset=['Customer ID'], keep='first',inplace=True)
print("\nAfter dropping duplicate Customer ID : {} values for Dataframe1".format(len(df1)))
print("After dropping duplicate Customer ID : {} values for Dataframe2".format(len(fake_df)))
```
